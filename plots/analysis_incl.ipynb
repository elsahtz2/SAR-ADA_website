```python
from dash import Dash, dcc, html, Input, Output, State
import base64
import matplotlib.colors as mcolors

# Initialize the Dash app
app = Dash(__name__)

def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode()

# Preload images
for mv_list in [movies_inclusivity, movies_success, movies_mean]:
    for m in mv_list:
        m["encoded_image"] = encode_image(m["image"])

def make_marks(values):
    return {val: {'label': f"{val}", 'style': {'fontSize': '14px', 'color': '#333'}} for val in values}

def get_dataset_and_metric(selected_metric):
    if selected_metric == "inclusivity_score":
        dataset = movies_inclusivity
        metric = "inclusivity_score"
    elif selected_metric == "success_score":
        dataset = movies_success
        metric = "success_score"
    else:
        for m in movies_mean:
            m['mean_success_inclusivity'] = round((m['success_score'] + m['inclusivity_score'])/2.0,1)
        dataset = movies_mean
        metric = "mean_success_inclusivity"
    return dataset, metric

app.layout = html.Div([
    html.Div("Click button to choose how to sort movies:", 
             style={'fontSize': '18px', 'color': '#333', 'textAlign': 'center', 'marginBottom': '10px'}),

    # Button to show/hide the dropdown
    html.Button("Sort movies by", id='show-options-button', n_clicks=0, 
                style={'margin': '10px auto', 'display': 'block', 'fontSize': '16px', 'padding': '10px 20px', 'borderRadius': '5px', 'cursor': 'pointer'}),

    # Store boolean for dropdown visibility
    dcc.Store(id='dropdown-visible', data=False),

    # Dropdown for metric selection, initially hidden
    html.Div([
        dcc.Dropdown(
            id='metric-dropdown',
            options=[
                {'label': 'Inclusivity Score', 'value': 'inclusivity_score'},
                {'label': 'Success Score', 'value': 'success_score'},
                {'label': 'Mean (Success & Inclusivity)', 'value': 'mean_success_inclusivity'}
            ],
            value='inclusivity_score',
            clearable=False,
            style={'width': '300px', 'margin': '10px auto'}
        )
    ], id='dropdown-container', style={'textAlign': 'center', 'display': 'none'}),

    html.Div([
        html.Div(id='slider-label', style={
            'fontSize': '18px', 
            'color': '#333', 
            'textAlign': 'center', 
            'marginBottom': '10px'
        }),
        dcc.Slider(
            id='movie-slider',
            min=0,
            max=1,
            step=None,
            value=0,
            tooltip={"placement": "top", "always_visible": True},
            marks={},
        ),
    ], style={
        'margin': '20px 40px',
        'background': '#f2f2f2',
        'padding': '20px',
        'borderRadius': '10px',
        'overflow': 'hidden'
    }),


    html.Div(id='movie-info', style={'padding': '20px', 'maxWidth': '1200px', 'margin': '20px auto'})
])

@app.callback(
    Output('dropdown-visible', 'data'),
    Input('show-options-button', 'n_clicks'),
    State('dropdown-visible', 'data')
)
def toggle_dropdown(n_clicks, currently_visible):
    if n_clicks is None:
        return currently_visible
    # Toggle visibility each click
    return not currently_visible

@app.callback(
    Output('dropdown-container', 'style'),
    Input('dropdown-visible', 'data')
)
def show_hide_dropdown(visible):
    return {'textAlign': 'center', 'display': 'block' if visible else 'none'}

@app.callback(
    [Output('movie-slider', 'min'),
     Output('movie-slider', 'max'),
     Output('movie-slider', 'marks'),
     Output('movie-slider', 'value'),
     Output('movie-info', 'children'),
     Output('slider-label', 'children')],
    [Input('metric-dropdown', 'value'),
     Input('movie-slider', 'value')]
)
def update_display(selected_metric, selected_value, good_bechdel=2):

    metric_label_map = {
        'inclusivity_score': 'Inclusivity Score',
        'success_score': 'Success Score',
        'mean_success_inclusivity': 'Mean (Success & Inclusivity)'
    }
    display_metric_name = metric_label_map[selected_metric]

    dataset, metric = get_dataset_and_metric(selected_metric)
    sorted_movies = sorted(dataset, key=lambda x: x[metric])
    metric_values = [m[metric] for m in sorted_movies]

    if selected_value not in metric_values:
        selected_value = metric_values[0]

    fontstyle = 'Open Sans'
    titlesize = '50px'
    headersize = '22px'
    textsize = '18px'
    

    def interpolate_green(value, min_val, max_val):
        norm = (value - min_val) / (max_val - min_val)
        norm = max(0, min(norm, 1))
        base_color = np.array(mcolors.to_rgb('#a7d8b1'))
        target_color = np.array(mcolors.to_rgb('#10a72f'))
        interpolated_rgb = base_color * (1 - norm) + target_color * norm
        return mcolors.to_hex(interpolated_rgb)

    def interpolate_red(value, min_val, max_val):
        norm = (value - min_val) / (max_val - min_val)
        norm = max(0, min(norm, 1))
        base_color = np.array(mcolors.to_rgb('#a40606'))
        target_color = np.array(mcolors.to_rgb('#dcbfbf'))
        interpolated_rgb = base_color * (1 - norm) + target_color * norm
        return mcolors.to_hex(interpolated_rgb)

    def score_color(score):
        return interpolate_green(score, 0, 1) if score >= 0.5 else interpolate_red(score, 0, 1)

    def bechdel_color(bechdel):
        return interpolate_green(bechdel, 0, 3) if bechdel >= good_bechdel else interpolate_red(bechdel, 0, 3)

    def revenue_color(revenue):
        boxoffice_min = -1e+06
        boxoffice_max = 1e+08
        return interpolate_green(revenue, boxoffice_min, boxoffice_max) if revenue >= 0 else interpolate_red(revenue, boxoffice_min, boxoffice_max)

    movie = next((m for m in sorted_movies if m[metric] == selected_value), sorted_movies[0])

    children = html.Div([
        # Left Side: Image
        html.Div([
            html.Img(src=f"data:image/jpeg;base64,{movie['encoded_image']}", 
                     style={'display': 'block', 'margin': 'auto', 'maxWidth': '350px', 'maxHeight': '450px'}),
            html.Br()
        ], style={'flex': '1', 'textAlign': 'center', 'padding': '10px'}),

        # Right Side: Text
        html.Div([
            html.H1(movie["title"], style={
                'textAlign': 'center',
                'marginBottom': '10px',
                'fontSize': titlesize,
                'fontFamily': fontstyle,
                'color': '#333333',
            }),
            html.P([
                html.Span("Year: ", style={'fontWeight': 'bold', 'fontFamily': fontstyle, 'fontSize': headersize, 'color': '#333333'}),
                f"{movie['year']}",
            ], style={'textAlign': 'left', 'marginTop': '15px', 'fontSize': textsize, 'fontFamily': fontstyle, 'color': '#333333'}),
            
            html.P([
                html.Span("Inclusivity Score: ", style={'fontWeight': 'bold', 'fontFamily': fontstyle, 'fontSize': textsize, 'color': '#333333'}),
                html.Span(f"{movie['inclusivity_score']}", style={'color': bechdel_color(movie['bechdel_score'])}),
                html.Br(),
                html.Span("Bechdel Score: ", style={'fontFamily': fontstyle, 'fontSize': textsize, 'color': '#333333'}),
                html.Span(f"{movie['bechdel_score']}", style={'color': bechdel_color(movie['bechdel_score'])}),
                html.Br(),
                html.Span("Proportion of Actress: ", style={'fontFamily': fontstyle, 'fontSize': textsize, 'color': '#333333'}),
                html.Span(f"{movie['proportion_of_actress']}", style={'color': score_color(movie['proportion_of_actress'])}),
                html.Br(),
                html.Span("Share of Women Script: ", style={'fontFamily': fontstyle, 'fontSize': textsize, 'color': '#333333'}),
                html.Span(f"{movie['share_of_women_script']}", style={'color': score_color(movie['share_of_women_script'])}),
            ], style={'textAlign': 'left', 'marginTop': '10px', 'fontSize': textsize, 'fontFamily': fontstyle, 'color': '#333333'}),

            html.P([
                html.Span("Success Score: ", style={'fontWeight': 'bold', 'fontFamily': fontstyle, 'fontSize': headersize, 'color': '#333333'}),
                html.Span(f"{movie['success_score']}", style={'color': score_color(movie['success_score'])}),
                html.Br(),
                html.Span("Imdb rating: ", style={'fontFamily': fontstyle, 'fontSize': textsize, 'color': '#333333'}),
                html.Span(f"{movie['Imdb_rating']}", style={'color': score_color(movie['Imdb_rating'])}),
                html.Br(),
                html.Span("Box Office Revenue: ", style={'fontFamily': fontstyle, 'fontSize': textsize, 'color': '#333333'}),
                html.Span(f"${int(movie['revenue']):,}", style={'color': revenue_color(movie['revenue'])}),
            ], style={'textAlign': 'left', 'marginTop': '10px', 'fontSize': textsize, 'fontFamily': fontstyle, 'color': '#333333'}),
            
            html.P([
                html.Span("Synopsis: ", style={'fontWeight': 'bold', 'fontSize': headersize, 'fontFamily': fontstyle, 'color': '#333333'}),
                movie['synopsis']
            ], style={'textAlign': 'left', 'marginTop': '5px', 'fontSize': textsize, 'fontFamily': fontstyle, 'color': '#333333'}),
        ], style={'flex': '1', 'textAlign': 'left', 'padding': '10px'}),
    ], style={'display': 'flex', 'backgroundColor': '#f9f9f9', 'padding': '20px', 'borderRadius': '10px', 'boxShadow': '0 4px 8px rgba(0, 0, 0, 0.1)'})

    new_marks = make_marks(metric_values)
    return min(metric_values), max(metric_values), make_marks(metric_values), selected_value, children, f"Select a Movie by {display_metric_name}:"

if __name__ == '__main__':
    app.run_server()
```
